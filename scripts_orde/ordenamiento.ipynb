```python
df.head()
```

## Eliminación de duplicados

```python
df.drop_duplicates(subset=["UserId", "ProductId", "Text"], inplace=True)
```

## Manejo de valores nulos


```python
print(df.isnull().sum())
```

## Rellenar valores faltantes en 'Summary' y 'ProfileName' con texto genérico


```python
df["Summary"].fillna("No summary", inplace=True)
df["ProfileName"].fillna("Unknown", inplace=True)
```

## Filtrar reseñas inconsistentes

```python
import numpy as np
df = df[df["HelpfulnessDenominator"] > 0]
# Quitar casos donde los votos útiles superan a los totales (datos inválidos)
df = df[df["HelpfulnessNumerator"] <= df["HelpfulnessDenominator"]]
# =====================================================
#  4. Convertir timestamp a fecha legible
# =====================================================
df["Time"] = pd.to_datetime(df["Time"], unit='s')

# =====================================================
#   Crear una variable de "utilidad" (target)
# =====================================================
# Ejemplo: útil si más del 60% de los votos la consideraron útil
df["HelpfulRatio"] = df["HelpfulnessNumerator"] / df["HelpfulnessDenominator"]
df["HelpfulLabel"] = np.where(df["HelpfulRatio"] >= 0.6, 1,0)

# =====================================================
#  6. Eliminar columnas innecesarias
# =====================================================
df_clean = df.drop(columns=["Id", "UserId", "ProfileName", "HelpfulnessNumerator",
                            "HelpfulnessDenominator", "HelpfulRatio"])
```

## guarde

```python
#   Guardar dataset limpio
# =====================================================
df_clean.to_csv("amazon_reviews_clean.csv", index=False)

print(" Limpieza completada. Dataset listo para análisis o modelado.")
print(df_clean.head())
```
